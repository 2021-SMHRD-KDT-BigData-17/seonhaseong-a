```python
y_encoded = tf.keras.utils.to_categorical(y_val)
print(y_encoded)
```

```python
X_train, X_test, y_train,y_test = train_test_split(X,y_encoded, test_size=0.2,random_state=5)
```

```python
from sklearn.model_selection import StratifiedKFold # 교차검증
n_fold = 10
seed =0
skf = StratifiedKFold(n_fold, shuffle=True,random_state=seed)
```

```python
from keras.models import Sequential
```

```python
accuracy =[]
for train, test in skf.split(X,y):
  model = Sequential()
  model.add(Dense(16, input_dim= 17, activation = 'relu'))
  model.add(Dense(32, activation='relu'))
  model.add(Dense(64, activation='relu'))
  model.add(Dense(128, activation='relu'))
  model.add(Dense(6, activation ='softmax'))

  model.compile(optimizer = 'adam', loss='categorical_crossentropy', metrics=['accuracy'])

  model.fit(X_train,y_train, epochs=200, batch_size=40,validation_data=(X_test,y_test))

  # 모델 예측
  k_accuracy = model.evaluate(X_test,y_test)[1]
  accuracy.append(k_accuracy)

print(accuracy)
```

```python
# 검증 셋의 오차값 저장
y_vloss=history.history['val_loss']
# 학습셋의 정확도 trainning  data
y_acc = history.history['accuracy']

# x값 지정 후 정확도를 파란색, 오차를 빨간색으로 표현 할거임
x_len =np.arange(len(y_acc))
plt.plot(x_len,y_vloss,'o', c='red', markersize=13)
plt.plot(x_len,y_acc,'o',c='blue', markersize=13)

plt.show()
```
